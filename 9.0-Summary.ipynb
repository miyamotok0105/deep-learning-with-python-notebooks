# ディープラーニングについてどう考えるべきか

モデルは非常に単純。ディープラーニングでは、全てがベクトル。全てが幾何学空間における1つの点。1つ1つの層で入力空間から目的空間へ１点ずつ写像（マッピング）される。この幾何学変換は微分可能でないとできない。

別の言い方をすると、くしゃくしゃに丸めた紙の玉が入力データの多様体。人が紙の玉を開いて行く動作の1つ1つが、1つの層によって実行される単純な幾何学変換。ディープラーニングは高次元データからなる複雑な多様体を解いて行く数学的機械。

ニューラルネットという名前は他の名前をつけるとしたら、階層化表現学習、階層的表現学習、ディープ微分可能モデル、連鎖幾何学変換
などという名前でもよかったかもしれない。

#絶え間ない学習とモジュール型のサブルーチンの再利用

学習済みの特徴量だけではなく、モデルのアーキテクチャや訓練手続きもプログラム言語の関数やクラスのように再利用するようになる。
現在は開発時に既存ライブラリを検索し、ダウンロードし利用します。同様に再利用可能な高度なブロックからなるグローバルライブラリの取捨選択により、メタ学習システムが新しいプログラムを構築できるようになる。
